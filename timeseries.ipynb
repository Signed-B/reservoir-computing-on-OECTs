```python
import shelve
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

#### 3D Time Series Plotting

```python
dimindex=1 # index in reservoir_dims of the dimension to plot

with shelve.open("./Data/dims_ensemble/signals_ensemble") as data:
    datadicts = data['dicts']
    reservoir_dims = data['dims']
    t = data['time']

OECT_signals = datadicts[0]["OECT_signals"][dimindex]
OECT_predictions = datadicts[0]["OECT_predictions"][dimindex]
tanh_signals = datadicts[0]["tanh_signals"][dimindex]
tanh_predictions = datadicts[0]["tanh_predictions"][dimindex]
```

##### Tanh plot

```python


plt.figure(figsize=(10, 10))

plot_index = 1

start = 0
end = 1000
end_3d = 10000

to_signals = tanh_signals
to_predictions = tanh_predictions

x_sig = [sig[0] for sig in to_signals]
y_sig = [sig[1] for sig in to_signals]
z_sig = [sig[2] for sig in to_signals]
x_pred = [pred[0] for pred in to_predictions]
y_pred = [pred[1] for pred in to_predictions]
z_pred = [pred[2] for pred in to_predictions]

print("Length of timeseries:", len(x_sig))


# x
plt.subplot(3, 1, plot_index)
plt.title("x")
plt.plot(t[start:end], x_sig[start:end], "k-", label="ground truth")
plt.plot(t[start:end], x_pred[start:end], "r-.", label="prediction")

if plot_index == 1:
    plt.legend()
plot_index += 1

# y
plt.subplot(3, 1, plot_index)
plt.title("y")
plt.plot(t[start:end], y_sig[start:end], "k-", label="ground truth")
plt.plot(t[start:end], y_pred[start:end], "r-.", label="prediction")
plot_index += 1

# z
plt.subplot(3, 1, plot_index)
plt.title("z")
plt.plot(t[start:end], z_sig[start:end], "k-", label="ground truth")
plt.plot(t[start:end], z_pred[start:end], "r-.", label="prediction")
plot_index += 1


plt.ylabel("Value")
plt.xlabel("time")
plt.show()
```

#### OECT Plot

```python
plt.cla()
plt.clf()

plt.figure(figsize=(10, 10))
plt.suptitle("OECT vs tanh timeseries")
plt.subplots_adjust(top=0.93, hspace=0.5, wspace=0.5)
plot_index = 1

start = 0
end = 1000
end_3d = 10000

to_signals = OECT_signals
to_OECT = OECT_predictions
to_tanh = tanh_predictions

x_sig = [sig[0] for sig in to_signals]
y_sig = [sig[1] for sig in to_signals]
z_sig = [sig[2] for sig in to_signals]
x_oect = [pred[0] for pred in to_OECT]
y_oect = [pred[1] for pred in to_OECT]
z_oect = [pred[2] for pred in to_OECT]
x_tanh = [pred[0] for pred in to_tanh]
y_tanh = [pred[1] for pred in to_tanh]
z_tanh = [pred[2] for pred in to_tanh]


print("Length of timeseries:", len(x_sig))


# x
plt.subplot(3, 1, plot_index)
plt.title("x")
plt.plot(t[start:end], x_sig[start:end], "k-", label="ground truth")
plt.plot(t[start:end], x_oect[start:end], "r--", label="OECT prediction")
plt.plot(t[start:end], x_tanh[start:end], "b-.", label="tanh prediction")

if plot_index == 1:
    plt.legend()
plot_index += 1

# y
plt.subplot(3, 1, plot_index)
plt.title("y")
plt.plot(t[start:end], y_sig[start:end], "k-", label="ground truth")
plt.plot(t[start:end], y_oect[start:end], "r--", label="OECT prediction")
plt.plot(t[start:end], y_tanh[start:end], "b-.", label="tanh prediction")
plot_index += 1

# z
plt.subplot(3, 1, plot_index)
plt.title("z")
plt.plot(t[start:end], z_sig[start:end], "k-", label="ground truth")
plt.plot(t[start:end], z_oect[start:end], "r--", label="OECT prediction")
plt.plot(t[start:end], z_tanh[start:end], "b-.", label="tanh prediction")
plot_index += 1


plt.ylabel("Value")
plt.xlabel("time")
plt.show()
```
